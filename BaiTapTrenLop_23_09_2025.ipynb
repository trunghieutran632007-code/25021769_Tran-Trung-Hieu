```python
#bai1: tam giac
a, b, c= input("Nhập 3 số: ").split()
a, b, c = int(a), int(b), int(c)
if (a+b<=c) or (a+c<=b) or (b+c<=a):
    print("Khong phai 3 canh cua tam giac")
else:
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    print(f"Diện tích tam giác là: {s:.1f}")
```
