```python
#Bai 3
a = int(input("Nhap so nguyen duong a: "))

if a > 0:
    print(f"BANG CUU CHUONG CUA {a}")
    for i in range(1, 11):
        print(f"{a} x {i} = {a * i}")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 4
h = int(input("Nhap chieu cao h: "))

for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print(" " * (h - i) + "*" + " " * (2 * i - 3) + "*")
```

```python
#Bai 5
n = int(input("Nhap so nguyen duong n: "))

if n > 0:
    S = 0
    for i in range(1, n + 1):
        S += i
    print("Tong S =", S)
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 6
a = int(input("Nhap so nguyen duong a: "))

if a > 0:
    print(f"Cac uoc duong cua {a} la:")
    for i in range(1, a + 1):
        if a % i == 0:
            print(i, end=" ")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 7
a = int(input("Nhap so nguyen duong a: "))
count = 0

if a > 0:
    for i in range(1, a + 1):
        if a % i == 0:
            count = count + 1
    print(f"So uoc duong cua {a} la: {count}")
    
            
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 8
a, b = map(int, input("Nhap 2 so nguyen duong a, b: ").split())

if a > 0 and b > 0:
    print(f"Cac uoc duong chung cua {a} va {b} la:")
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
else:
    print("Vui long nhap 2 so nguyen duong!")
```
