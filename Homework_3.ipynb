```python
#Bai 3
a = int(input("Nhap so nguyen duong a: "))

if a > 0:
    print(f"BANG CUU CHUONG CUA {a}")
    for i in range(1, 11):
        print(f"{a} x {i} = {a * i}")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 4
h = int(input("Nhap chieu cao h: "))

for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print(" " * (h - i) + "*" + " " * (2 * i - 3) + "*")
```

```python
#Bai 5
n = int(input("Nhap so nguyen duong n: "))

if n > 0:
    S = 0
    for i in range(1, n + 1):
        S += i
    print("Tong S =", S)
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 8
a, b = map(int, input("Nhap 2 so nguyen duong a, b: ").split())

if a > 0 and b > 0:
    print(f"Cac uoc duong chung cua {a} va {b} la:")
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
else:
    print("Vui long nhap 2 so nguyen duong!")
```

```python
#Bai 9
a = int(input("Nhap so nguyen duong a: "))

if a > 0:
    if int(a ** 0.5) ** 2 == a:
        print(f"{a} la so chinh phuong")
    else:
        print(f"{a} khong phai la so chinh phuong.")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 10
a = int(input("Nhap so nguyen duong a: "))

if a > 0:
    tong = 0
    for i in range(1, a):
        if a % i == 0:
            tong += i
    if tong == a:
        print(f"{a} la so hoan hao")
    else:
        print(f"{a} khong phai la so hoan hao")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 12
n = int(input("Nhap so nguyen duong n: "))

if n > 0:
    k = 0
    S = 0
    while True:
        k += 1
        S += k
        if S >= n:
            k -= 1   #Ta cho chay S(k) tu 1->k, neu S(k)>=n thi ta se cho k quay lai gia tri truoc do
            break
    print(f"k lon nhat la {k}")
    print(f"S({k}) = {k * (k + 1) // 2}") #Cong thuc tinh tong day 1+2+...+k
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 13
A = float(input("Nhap so nguyen duong A: "))

if A > 0:
    S = 0
    n = 0
    while S <= A:
        n += 1
        S += 1 / n
    print(f"n nho nhat: {n}")
    print(f"Tong S = {S}")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 14
n = int(input("Nhap mot so tu nhien: "))
max = n
min = n
while True:
    if n != -1:
        n = int(input("Nhap mot so tu nhien: "))
        if n>max:
            max = n
        if n<min:
            min = n
    else:
        print(f'So lon nhat la: {max}, So be nhat la: {min}')
        break
        
```
