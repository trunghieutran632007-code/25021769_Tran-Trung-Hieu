```python
#Bai 4
h = int(input("Nhap chieu cao h: "))

for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print(" " * (h - i) + "*" + " " * (2 * i - 3) + "*")
```

```python
#Bai 12
n = int(input("Nhap so nguyen duong n: "))

if n > 0:
    k = 0
    S = 0
    while True:
        k += 1
        S += k
        if S >= n:
            k -= 1   #Ta cho chay S(k) tu 1->k, neu S(k)>=n thi ta se cho k quay lai gia tri truoc do
            break
    print(f"k lon nhat la {k}")
    print(f"S({k}) = {k * (k + 1) // 2}") #Cong thuc tinh tong day 1+2+...+k
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 13
A = float(input("Nhap so nguyen duong A: "))

if A > 0:
    S = 0
    n = 0
    while S <= A:
        n += 1
        S += 1 / n
    print(f"n nho nhat: {n}")
    print(f"Tong S = {S}")
else:
    print("Vui long nhap so nguyen duong!")
```

```python
#Bai 15_3 cach
import math

def count_digits_str(n):
    """Convert to string and count characters (handles n==0)."""
    n = abs(int(n))
    return len(str(n))

def count_digits_while(n):
    """Divide by 10 repeatedly."""
    n = abs(int(n))
    if n == 0:
        return 1
    cnt = 0
    while n:
        cnt += 1
        n //= 10
    return cnt

def count_digits_log(n):
    """Use log10. Works for n>0. Handle n==0 separately."""
    n = abs(int(n))
    if n == 0:
        return 1
    return math.floor(math.log10(n)) + 1
if __name__ == "__main__":
    try:
        s = input("Nhap vao mot so nguyen duong n: ").strip()
        n = int(s)
        if n < 0:
            print("Da doi sang so duong (abs).")
        print("Results:")
        print("1) str + len        :", count_digits_str(n))
        print("2) while //10       :", count_digits_while(n))
        print("3) math.log10       :", count_digits_log(n))
        
    except ValueError:
        print("Nhap khong hop le. Vui long nhap mot so nguyen.")
```

```python
#Bai 16
n = int(input("Nhap so nguyen duong n: "))
chan = le = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10

print("So chu so chan:", chan)
print("So chu so le  :", le)
```

```python
#Bai 18
n = int(input("Nhap so nguyen duong n: "))

if n < 1:
    print("Nhap so nguyen duong lon hon 0.")
else:
    while n % 3 == 0:
        n //= 3
    if n == 1:
        print("So co dang 3^k")
    else:
        print("So khong co dang 3^k")
```

```python
#Bai 19
A = int(input("Nhap so nguyen duong A: "))

f1, f2 = 1, 1  # hai so dau tien trong day Fibonacci
while f2 <= A:
    f1, f2 = f2, f1 + f2

print("So Fibonacci lon nhat khong vuot qua A la:", f1)
```

```python
#Bai 20
n = int(input("Nhap so nguyen duong n: "))

count = 0
temp = n  

while n > 0:
    d = n % 10      
    if d > 1:       
        is_prime = True
        for i in range(2, int(d**0.5) + 1):
            if d % i == 0:
                is_prime = False
                break
        if is_prime:
            count += 1
    n //= 10

print(f"So {temp} co {count} chu so la so nguyen to")
```
