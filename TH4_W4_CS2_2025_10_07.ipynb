```python
#W4A1
try:
    n = int(input('Nhap so nguyen n lon hon 1: '))
    if (n <= 0):
        print('Ban phai nhap vao mot so nguyen lon hon 1!')
    else:
        
        sum = 0
        for i in range(1, n+1):
            sum += i

        print(f'Tong cac so tu 1 den {n} la: {sum}')
except ValueError:
    print('Gia tri nhap vao khong hop le')

```

```python
#W4A2
try:
    n = int(input('Nhap so nguyen n lon hon 1: '))
    if (n <= 0):
        print('Ban phai nhap vao mot so nguyen duong (>0)!')
    else:
        print(f'Ban nhap vao {n}')
        if (n == 1):
            print('1 khong phai so nguyen to')
        else:
            nguyenTo = True
            for i in range(2, int(n**0.5) + 1):
                if n % i == 0:
                    nguyenTo = False
                    break
            if (nguyenTo):
                print(f'{n} la so nguyen to')
            else:
                print(f'{n} khong la so nguyen to')
except ValueError:
    print('Gia tri khong hop le')



```

```python
#W4A5
while True:
    try:
        flag = False
        n = int(input('Nhap so nguyen n'))
        if (n > 0):
            for _ in range(n):
                x = int(input())
                if (x == 42):
                    flag = True
            if (flag):
                print("I've found meaning of life")
            else:
                print("It's a joke")
            break
        else:
            print('Gia tri khong hop le')
    except ValueError:
        print('Gia tri khong hop le')
```

```python
#W4A6
tong = 0
while True:
    try:
        a, b = map(int, input('Nhap 2 so a, b cho doan [a,b]: ').split())
        if a <= b:
            for i in range(a, b + 1):
                if i > 1:  # loại bỏ 0 và 1
                    flag = True
                    for k in range(2, int(i**0.5) + 1):
                        if i % k == 0:
                            flag = False
                            break  # có ước rồi thì thoát sớm
                    if flag:
                        tong += i

            print(f'Tong cac so nguyen to trong doan [{a},{b}] la: {tong}')
            break
        else:
            print('Gia tri khong hop le! a phai <= b')

    except ValueError:
        print('Gia tri khong hop le!')
```

```python
#W4A8
n = int(input("Nhap so nguyen duong n: "))
reverse = str("")
check = 0

while check == 0:
    tmp = n
    reverse = ""
    while tmp > 0:
        reverse += str(tmp % 10)
        tmp //= 10

    m = 0
    for x in reverse:
        m = m * 10 + int(x)
    n += m

    s = str(n)
    check = 1
    for i in range(0, len(s) // 2):
        if s[i] != s[len(s) - 1 - i]:
            check = 0

print(f'So can tim la {n}')
```
