```python
#Bai 5
a, b, c = map(float, input('Nhap 3 so a,b,c cua pt bac 2 ax^2+bx+c=0: ').split())
if(a==0):
  if(b==0):
    if(c==0):
      print('Phuong trinh vo so nghiem')
    else:
      print('Phuong trinh vo nghiem')
  else:
    print(f'Nghiem cua phuong trinh la: {-c/b}')
else:
  delta = b*b-4*a*c
  if(delta<0):
    print('Phuong trinh vo nghiem')
  elif(delta==0):
    print(f'Phuong trinh co nghiem kep la: {-b/(2*a)}')
  else:
    print(f'Phuong trinh co 2 nghiem phan biet la: {(-b+delta**0.5)/(2*a)} va {(-b-delta**0.5)/(2*a)}')
```

```python
#Bài 8
print("Giải hệ phương trình: ")
print("a*x + b*y = m")
print("c*x + d*y = n")

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm.")
    else:
        print("Hệ phương trình vô nghiệm.")
else:
    x = Dx / D
    y = Dy / D
    print(f"Nghiệm của hệ là: x = {x:.2f}, y = {y:.2f}")
```
