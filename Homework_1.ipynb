```python
#Bài8
# Nhập dữ liệu
ho_ten = input("Nhập họ tên chủ hộ: ")
cs_truoc = int(input("Nhập chỉ số điện kế tháng trước: "))
cs_sau = int(input("Nhập chỉ số điện kế tháng này: "))

# Tính số kWh tiêu thụ
so_kwh = cs_sau - cs_truoc

# Đơn giá theo bậc thang
gia_bac = [1984, 2050, 2380, 2998, 3350, 3460]

# Hàm tính tiền điện theo bậc
def tinh_tien_dien(kwh):
    tien = 0
    if kwh <= 50:
        tien = kwh * gia_bac[0]
    elif kwh <= 100:
        tien = 50 * gia_bac[0] + (kwh - 50) * gia_bac[1]
    elif kwh <= 200:
        tien = 50 * gia_bac[0] + 50 * gia_bac[1] + (kwh - 100) * gia_bac[2]
    elif kwh <= 300:
        tien = 50 * gia_bac[0] + 50 * gia_bac[1] + 100 * gia_bac[2] + (kwh - 200) * gia_bac[3]
    elif kwh <= 400:
        tien = 50 * gia_bac[0] + 50 * gia_bac[1] + 100 * gia_bac[2] + 100 * gia_bac[3] + (kwh - 300) * gia_bac[4]
    else:
        tien = 50 * gia_bac[0] + 50 * gia_bac[1] + 100 * gia_bac[2] + 100 * gia_bac[3] + 100 * gia_bac[4] + (kwh - 400) * gia_bac[5]
    return tien

# Tính tiền trước VAT
tien_truoc_vat = tinh_tien_dien(so_kwh)

# VAT 8%
vat = tien_truoc_vat * 0.08

# Tổng tiền
tong_tien = round(tien_truoc_vat + vat)

# Xuất kết quả
print(f"\nChủ hộ: {ho_ten}")
print(f"Số điện tiêu thụ: {so_kwh} kWh")
print(f"Tiền điện (chưa VAT): {tien_truoc_vat} đồng")
print(f"VAT (8%): {int(vat)} đồng")
print(f"Tổng tiền phải trả: {tong_tien} đồng")
```
